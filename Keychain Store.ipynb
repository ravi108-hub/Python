#problem. write a program that displays below menu after every single choice given until checkout. do operations as per the menu, add keychains, remove, view order and checkout.
1) add keychains to order
2) remove keychains from order
3) view current order
4) checkout

```python
counter=0
for i in range(100): 

    def addKeyChain(keyChain):
        global counter
        counter=counter+keyChain
        #print("counter:" ,counter)
        #print("keyChain: ",keyChain)
        return counter

    def remKeyChain(keyChain):
        global counter
        counter=counter-keyChain
        return counter
        

    def viewOrder():
        global counter
        print("Each keyChain cost is $10")
        print("your total bucket cost is: $",counter*10)
        print("do you wish to checkout!!!")
        
    print(" ")
    print("1.add keychains to order ")
    print("2.remove keychains from order")
    print("3.view current order")
    print("4.checkout")
    print(" ")
    n=input("Enter your choice:  ")
    print(" ")

    if n == '1':
        print("you have" , counter , "keychains")
        keyChain=int(input("How many to add: "))
        result=addKeyChain(keyChain)                    #in result it will take output of f() i.e. return counter
        print("you now have: " ,result,"keychains")
        print(" ")

    elif n == '2':
        print("you have" , counter , "keychains")
        keyChain=int(input("How many to remove: "))
        result=remKeyChain(keyChain)
        print("you now have: " ,result,"keychains")
        print(" ")
        
    elif n == '3':
        print("you have", counter , "keychains in your bucket")
        viewOrder()
        print(" ")
        
    elif n == '4':
        print("You are about to Checkout, please pay your Bill")
        print("Thanks for the shopping, Visit Again!!!")
        print(" ")
        break
        
    else:
        print("wrong input, please select from Menu below: ")
        print(" ")
```
